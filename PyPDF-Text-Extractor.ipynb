```python
#importing PyPDF 
from PyPDF2 import PdfReader
```

```python
# Importing and Reading PDF file
X = PdfReader("ISMS.pdf")

# Parsing PDF file pages 
Y = X.pages[22]

#Extracting text from PDF page
Text = Y.extract_text()
```
